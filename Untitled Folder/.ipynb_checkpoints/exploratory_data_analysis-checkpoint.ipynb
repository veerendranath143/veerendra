# Median absolute deviation
### does not square but takes absolute value without taking negative

```python
from statsmodels import robust
print("Mean Absolute Deviation")
print(robust.mad(iris_setosa['petal.length']))
print(robust.mad(iris_virginica['petal.length']))
print(robust.mad(iris_versicolor['petal.length']))
```

##### Median -> Mean
##### Mad -> std - dev
##### IQR-> inter quartile range

# cumulative distribute function
# probability density function

```python
counts,bin_edges= np.histogram(iris_setosa['petal.length'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

#### blue line is called pdf
#### orange line is called cdf
### cdfs is sum of pdfs is called cumulative frequency
#### differenciation of cdf is called pdf
#### integration of pdf is called cdf

```python
counts,bin_edges= np.histogram(iris_setosa['petal.length'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges= np.histogram(iris_virginica['petal.length'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts,bin_edges= np.histogram(iris_versicolor['petal.length'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
